```python
class color:
   BOLD = '\033[1m'
   END = '\033[0m'

import requests
from bs4 import BeautifulSoup
import pandas as pd 
import json
```

```python
streamServices= {}
streamServices[1]="iTunes"
streamServices[3]="Maxdome"
streamServices[4]="Google Play"
streamServices[7]="Videoload"
streamServices[8]="Netzkino"
streamServices[10]="Amazon"
streamServices[11]="Netflix"
streamServices[12]="Sky Go"
streamServices[13]="Microsoft"
streamServices[14]="Videobuster"
streamServices[15]="videociety"
streamServices[16]="Sony"
streamServices[17]="Mubi"
streamServices[18]="Sky Ticket"
streamServices[19]="realeyz"
streamServices[20]="Rakuten TV"
streamServices[21]="alleskino"
streamServices[22]="kividoo"
streamServices[23]="CHILI"
streamServices[24]="Flimmit"
streamServices[27]="FilmConfect"
streamServices[29]="TVNOW"
streamServices[30]="Sky Store"
streamServices[31]="Pantaflix"
streamServices[33]="Anime On Demand"
streamServices[34]="MagentaTV"
streamServices[35]="freenet Video"
streamServices[36]="Joyn"
streamServices[39]="Apple TV+"
streamServices[40]="filmingo"
streamServices[41]="Watch4"
```

```python
data = pd.read_csv("WATCHLIST.csv", encoding = "ISO-8859-1")
foundCounter = 0
streamCounter= {}
# for every entry in the watchlist
for i in range(0,data.count()[1]):
    movieTitle=data.loc[i][['Title']][0]
    movieYear=data.loc[i][['Year']][0].astype(int)
    URL = 'https://www.werstreamt.es/filme/?q='+movieTitle+""
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, 'html.parser')
    movies = soup.find_all('li',itemprop='itemListElement')
    #go over every found movie-entry on the website
    for movie in movies:
        if None in movie:
            continue
        metaData = movie.find('meta',itemprop='dateCreated')
        if metaData==None:
            continue
        if (str(movieYear) in metaData["content"]):
            availableAt=""
            streams = json.loads(movie['data-avastatus'])
            if streams!=[]:
                for key, value in streams.items():
                    if value:
                        for serviceID in streamServices.keys():
                            if key==str(serviceID):
                                availableAt +=streamServices[serviceID] +", "
                                if not serviceID in streamCounter:
                                    streamCounter[serviceID]=0
                                streamCounter[serviceID] += 1
            if availableAt!="" :
                print(color.BOLD+"("+str(movieYear)+")\t"+ movieTitle+color.END+"\t"+availableAt, end='\n')
                foundCounter+=1
                break
```

```python
print("Movies found: "+str(foundCounter)+" of "+str(data.count()[1]))

for serviceIDs in streamCounter.keys():
    print(streamServices[serviceIDs]+": "+str(streamCounter[serviceIDs]))
```
